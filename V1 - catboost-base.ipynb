```python
train = features_df.copy()
labels = labels_df.copy()
test = test_features_df.copy()
```

```python
for col in num_cols:
    train[col] = train[col].fillna(value=-1)

for col in (cat_cols):
    train[col] = train[col].fillna(value='missing')
```

```python
for col in num_cols:
    test[col] = test[col].fillna(value=-1)

for col in (cat_cols):
    test[col] = test[col].fillna(value='missing')
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
labels.status_group.unique()
```

```python
oversample = RandomOverSampler(sampling_strategy={'functional': 22568, 'non functional': 16000, 'functional needs repair': 8000})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( train, labels, test_size=0.3, random_state=1)
```

```python
oversampled_X_train, oversampled_y_train = oversample.fit_resample(X_train, y_train)
```

```python
categorical_features_indices = np.where(X_train.dtypes != float)[0]
```

```python
from catboost import CatBoostClassifier
```

```python
final_model = CatBoostClassifier(
    n_estimators=100,
    verbose=False,  
    cat_features=categorical_features_indices,
    learning_rate=0.001,
    task_type='GPU'
    )
```

```python
grid = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}
grid_search_result = final_model.grid_search(grid, 
                                       X=oversampled_X_train, 
                                       y=oversampled_y_train, 
                                       plot=True)
```
