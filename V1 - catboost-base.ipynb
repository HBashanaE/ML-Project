```python
num_cols = features_df.select_dtypes('number').columns
cat_cols = features_df.select_dtypes('object').columns
# cat_cols = ['race', 'sex', 
#        'marital_status', 'rent_or_own',  'hhs_geo_region',
#        'census_msa', 'employment_industry', 'employment_occupation']

# ord_cols = ['age_group', 'education',  'income_poverty',
#         'employment_status']
```

```python
train = features_df.copy()
labels = labels_df.copy()
test = test_features_df.copy()
```

```python
for col in num_cols:
    train[col] = train[col].fillna(value=-1)

for col in (cat_cols):
    train[col] = train[col].fillna(value='missing')
```

```python
for col in num_cols:
    test[col] = test[col].fillna(value=-1)

for col in (cat_cols):
    test[col] = test[col].fillna(value='missing')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( train, labels, test_size=0.3, random_state=1)
```

```python
categorical_features_indices = np.where(X_train.dtypes != float)[0]
```

```python
from catboost import CatBoostClassifier
```

```python
final_model = CatBoostClassifier(verbose=False,  cat_features=categorical_features_indices, task_type='GPU')
```

```python
final_model.fit(X_train, y_train)
```

```python
predictions = final_model.predict(X_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
predictions[:, 0]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(predictions, y_test))
```
